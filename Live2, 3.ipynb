```python
#solving the above problem using recursion, refer image 3
#to find a permutation of size 3, use a for loop to get the 1st element and fix that element
#then apply a permutation of size 2 to rest of the 2 elements
#recursion will be inside the for loop
```

```python
# source: https://stackoverflow.com/questions/13109274/python-recursion-permutations
#refer image 4 for explaination

def permutation(s):
    
   if len(s) == 1: #base case
     return [s]

   perm_list = [] # resulting list
   for a in s:
     remaining_elements = [x for x in s if x != a]
     z = permutation(remaining_elements) # permutations of sub-list

     for t in z:
       perm_list.append([a] + t)

   return perm_list

s = [1,2,3];
p = permutation(s);
print(p)
```

```python
Logic to compute interval b/w a to b:
1. We take a from the initial or lower limit and compute f(a) ie sin(x)/x
2. We take a small width(delta), by breaking the x axis b/w a to b in n such intervals, then the width of each interval will be (b-a)/n
3. Call each such interval as delta
we know the area of a rectangle = height * width
we keep the delta very small
4. we know the height of x value here which is f(a) and width is delta, therefore f(a)*delta, is the area of 1st rectangle
5. compute the 2nd rectangle , x value is a + delta, therefore area of 2nd rectangle will be f(a + delta) * delta
6. compute the 3rd rectangle , x value is a + 2*delta, therefore area of 3rd rectangle will be f(a + 2*delta) * delta
7. sum up all the rectangles to get the whole area from a to b
we are missing the yellow spaces in the diagram while calculation of the rectangles, but we are also including some space once the peak is crossed
therefore the area is approximate
to compute distance which is close to the actual distance, we need to reduce the value of delta, or make the n larger, which will make the width smaller, this will reduce the missed out area and extra covered area
```

```python
Trapezoidal rule says that we use the rectangle + the above triangle to calculate the area, this will be a better approximation than just the rectangle
Trapezoid is a better approximation than a rectangle

To calculate the area of a trapezoid,
we know f(a), the point on y axis where x=a
we know f(a+delta), the point on y axis where x=x1

area of trapezoid = area of rectangle + area of upper triangle

area of rectangle = length * breadth
length = f(a)
breadth = delta
area of rectangle = f(a)*delta

area of triangle = 1/2 * base * height
base = delta
height = f(a+delta) - f(a)
area of triangle = 1/2 * delta * (f(a+delta) - f(a))

area of trapezoid = f(a)*delta + 1/2 * delta * (f(a+delta) - f(a))

refer image 5
```

```python
- Maxima & derivative
We can find the maxima(or minima) by computing the derivative of sin(x)/x and equate that with 0, we will get 2 values - minima and maxima, but since sin(x)/x b/w -2 and 2 has only a maxima, we will get only 1 value
```

```python
- What is a derivative intuitively? [SLOPE]
eg: d f(x)/dx at point x=-2 notation

derivative is the slope of the line 
draw a tangent to the point at y axis corresponding to x=-2
calculate the slope of the line = tan theta (tan theta = opp/adj) ie rate of change of y wrt x
refer image 6

df/dx = slope of f wrt x
slope = rate of change

To know if the slope is increasing/decreasing, we need to check its state wrt the increasing value of x-axis

The function is increasing at x=-2 to x=0 - as the value of x increases, that means the slope is positive, also tan theta is lesser than 90 degree
similarly at x=2 and as the value of x increases, the function is decreasing, therefore slope is negative, also tan theta is greater than 90 degree
refer image 7

slope = df/dx = (f(x+deltax) - f(x))/deltax as lim deltax -> 0 (infinitesimal change)
refer image 8

Slope at the maxima or optima is 0
wherever df/dx = 0, we say that the point is maxima or minima

if function is moving from positive to negative, then we get a maxima
if function is moving from negative to positive , then we get a minima
refer image 9

move from -2 to 2, compute the slope from -2 to 2 and check if the slope is close to 0, if it is then we got the maxima
```

```python

#- How to compute derivative/slope?
# f(x) = sin(x)/x
import math
def f(x):
    if x != 0:
        return math.sin(x)/x;
    else:
        return math.nan


# return f'(x) = derivative of f(x) at x without computing the derivative explicitly
def slopeF(x): 
    delta = 0.0001;
    
    return (f(x+delta) - f(x))/delta ; #slope = df/dx = (f(x+deltax) - f(x))/deltax as lim deltax -> 0 (infinitesimal change)

print(slopeF(0)); # test-cases??
```

```python
Problem-8: Estimate the value of PI
You can only use the formula for area and circumference of a circle in terms of PI.
IBM Research interview question

area=pie*r^2
circumference=2*pie*r

Pie was defined in the ancient world as circumference/2*radius(diameter) = c/2*r
```

```python
Approach: Monte Carlo Approach
permutation testing is an eg of  Monte Carlo Approach

refer image 16

Here we calculate pie using the ratio of Area of Square and Area of circle
Simulation Approach:
Next, we can throw darts(arrows), such that all arrows lie within the square, there will be several arrows within the circle too
In this, we uniformly generate random nos from 0,1 (square size), say we generate n random nos or darts where n=10000
Say out of these m darts fall into the circle
To find out if the dart is within the circle, we can check the distance of the dart with the centre and compare it with the radius, if it is less than radius, then it lies within the circle
We know pie/4:1 is same as m:n
Therefore we can calculate the pie value very easily
refer image 17
```

```python
# Area of square = 1
# Area of circle = PI * 0.5^2 = PI/4

# Refernce: https://en.wikipedia.org/wiki/Approximations_of_%CF%80 [Many many algorithms]

# Lets now use randomization  [a.k.a Monte carlo simualtion]
# Permutation-tets in HYpothesis tetsing in our AI Course is another example of Monte-Carlo simualtions
# reading: https://en.wikipedia.org/wiki/Monte_Carlo_method

import random
import math

def inCircle(x , y):
    if  math.sqrt((x-0.5)**2 + (y-0.5)**2) <= 0.5:
        return True;
    else:
        return False;


cntInCircle = 0; #m
n = 10000000;

for i in range(n):
    x = random.random();
    y = random.random();
    
    if inCircle(x,y):
        cntInCircle += 1;
        
print (cntInCircle/n * 4); # m/n = pie/4, therefore 4 * m/n = pie
```
